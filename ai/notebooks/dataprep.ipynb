# Sample Data
Generate sample data using a Bing query for images. [Sample Structure](https://gist.github.com/fchollet/0830affa1f7f19fd47b06d4cf89ed44d)

```python
import os
import shutil
from bingimages import bingimages as bi

def clear(base_dir):
    if os.path.exists(base_dir):
        shutil.rmtree(base_dir) 
    os.makedirs(base_dir)
    
def generate_data(base_dir, tags, count, data_type):
    for tag in tags:
        path = os.path.join(base_dir, data_type, tag.replace(" ", "_"))
        if not os.path.exists(path):
            os.makedirs(path)
        bi.fetch_images_from_keyword(tag, path, "", limit=count)

base_dir = 'data/catsndogs'
clear(base_dir)
```

```python
tags = ['cat', 'dog']

generate_data(base_dir, tags, 500, 'train')
generate_data(base_dir, tags, 100, 'validation')
```
